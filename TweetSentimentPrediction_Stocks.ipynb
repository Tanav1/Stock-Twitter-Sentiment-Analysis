```python
def get_tweets(query, count = 10): 

      tweets = [] 

      try: 
          # call twitter api to fetch tweets 
          fetched_tweets = api.search(q = query, count = count) 
          for tweet in fetched_tweets:
            # cleaning the tweets
            print(tweet)
            tweet= tweet.text
            tweet = tweet.lower()
            tweet = re.sub(r"[^\w$#\s]+", "", tweet)

            tokens = tweet.split()

            tokens = [token for token in tweet if not token.startswith('https')]
            output_string = ''.join(tokens)


# Print the result
            print(output_string)


# Print the result
            # print(tokens)



            # print(tweet)

            # getting the sentiment from textblob
            analysis = TextBlob(tweet)
            #Removing Punctuations, Numbers, and Special Characters

            senti= analysis.sentiment.polarity
            # labeling the sentiment
            if senti<0 :
              emotion = "NEG"
            elif senti>0:
              emotion= "POS"
            else:
              emotion= "NEU"
            # appending all data
            tweets.append((tweet, senti, emotion))
            
          return tweets
      except tweepy.TweepError as e:
          # print error (if any) 
          print("Error : " + str(e))
```
